1. IMPORTS

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

2. LOAD DATA

```python
df = pd.read_csv(
    r"E:/ALL Documents/LEVEL 6 Completed/Projects/Week 1 AI & ML & Linux/end-to-end-explainable-ai-system/data/processed/cleaned_data.csv"
)

X = df.drop(columns=["Loan_Status"])
y = df["Loan_Status"]
```

3. Train model

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

4. Generate recommendations

```python
df_test = X.copy()
df_test["Approval_Probability"] = model.predict_proba(X_scaled)[:, 1]
df_test["Recommendation"] = np.where(
    df_test["Approval_Probability"] >= 0.5,
    "Approve Loan",
    "Reject Loan"
)

df_test.head()
```

5. Business logic

The recommendation system uses predicted probability from the supervised learning model.

If approval probability â‰¥ 50%, the system recommends loan approval; otherwise, rejection.

This approach ensures consistency, transparency, and explainability in decision-making.
