This step introduces rule-based and heuristic reasoning to support loan decisions before machine learning.

1. Define heuristic rules

Heuristic Rules

- Higher income vs loan amount → safer approval
- Valid credit history → lower risk
- Missing credit history → uncertainty penalty

```python
import pandas as pd

df = pd.read_csv(
    r"E:\ALL Documents\LEVEL 6 Completed\Projects\Week 1 AI & ML & Linux\end-to-end-explainable-ai-system\data\raw\loan-prediction-dataset.csv"
)

df.head()
```

2. Simple rule scoring

```python
def risk_score(row):
    score = 0
    if row["Credit_History"] == 1:
        score += 2
    if row["LoanAmount"] < (row["ApplicantIncome"] / 10):
        score += 1
    return score
```

Calling the function directly

```python
risk_score(df.iloc[0])
```

Apply it to your dataset

```python
df["Risk_Score"] = df.apply(risk_score, axis=1)
df[["Credit_History", "ApplicantIncome", "LoanAmount", "Risk_Score"]].head()
```
